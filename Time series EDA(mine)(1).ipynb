# Time series EDA

1. Upward Trend
2. Stationary Data
3. Downward trend
4. Cyclic data


1. Upward trend
- mean will be increased overtime. 
- less variance
- Sales of AC during summers. This can be considered as cyclic data as it increases on summer and decreases on winters. Also if we consider, price inflation: we can say that everyone will buy AC one point or another


2. Downward trend
- Ex; price goes down by time. ex: very bad stock price at one course of time. 
- Mean will be decreased overtime.

3. Stationary
- Mean is same over the time
- Easy to forecast

4. Cyclic data
- Stocks. people sentiments
- Very difficult to forecast 
- Even sales data is also cyclic data and we can forecast easily when compared to stocl cyclic data

## Time resampling

```python
df_tesla.head()
```

```python
#resampling
 
df_tesla.resample(rule= "A") # rule A means: end of year : A stands for annual

# rules: check https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

```

```python
df_tesla.resample(rule= "A").min()

#min() in resamples takes the last day of the year and shows us the minimum in columns "High", "Low", "Open" etc
```

```python
type(df_tesla.resample(rule= "A").min()) # dtype is dataframe
```

```python
df_tesla.resample(rule= "A").min()['Low'].plot()

# over 5 years we can see that the minimum amount has been raised. 
```

```python
df_tesla.resample(rule= "A").min()['High'].plot()
```

```python
# quarterly 

df_tesla.resample(rule= "QS").max()
```

```python
df_tesla.resample(rule= "QS").max()["High"].plot()
```

```python
df_tesla.resample(rule= "BA").max() #business year
```

```python
df_tesla.resample(rule= "BQS").max() #business quarters
```

```python
df_tesla['Open'].resample(rule= "A").mean().plot(kind= 'bar')

```

```python
df_tesla['Open'].resample(rule= "M").mean().plot(kind= 'bar',figsize=(15,6)) # M-monthly
```

```python
# rolling()- Provides rolling window calculations over the underlying data in the given Series object
df_tesla["High"].rolling(10) 
#10 axis=0 (rows)
```

```python
df_tesla["High"].rolling(10).mean()
# first calculate mean of 1 n 2nd, then calculate result of this with3rd. go ahead till 10th
```

```python
df_tesla["High"].rolling(11).mean().head(20)
# rolling is 11, then 1st 10 will be nan. rolling =12 thn 1st 11 will be NaN
```

```python
df_tesla["High"].rolling(11).sum().head(20)
```

```python
#new col as Open 30 days rolling
df_tesla['Open: 30 days rolling']= df_tesla['Open'].rolling(30).mean()
df_tesla.head(31)
```

```python
#plotting Open and Open days rolling

df_tesla[['Open','Open: 30 days rolling']].plot(figsize=(12,5))

#Orange is rolling() with rolling, I can smoothen the curve. 
```
